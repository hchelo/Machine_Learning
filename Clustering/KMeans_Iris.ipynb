```python
# Edited by MSaavedra
# k-means for IrisDataSet
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

import numpy as np
import matplotlib.pyplot as plt

# Creamos datos aleatorios
#X = np.random.rand(100, 2)
iris = load_iris()
X = iris.data
plt.figure(1)
plt.scatter(X[:, 0], X[:, 1])

```

```python
# Creamos un objeto de KMeans con 3 clusters
kmeans = KMeans(n_clusters=3)

# Ejecutamos el algoritmo K-means
kmeans.fit(X)

# Obtenemos los centroides de los clusters
centroids = kmeans.cluster_centers_

# Obtenemos las etiquetas de los clusters para cada muestra
labels = kmeans.labels_

# Visualizamos los resultados
plt.figure(2)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300, c='r')
plt.show()
```
